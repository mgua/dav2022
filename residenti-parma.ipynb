```python
import pandas as pd
import matplotlib.pyplot as plt
rp = pd.read_csv('residenti-parma.csv', delimiter=';' )
rp.head(5) 
```

```python
rp.info()
```

```python
rp.describe()
```

```python
rp.isnull().sum()
```

```python
rp['Sesso'].max()
```

```python
femmine = rp[rp['Sesso']==2]
maschi = rp[rp['Sesso']==1]
```

```python
femmine
```

```python
maschi
```

```python
maschi.groupby(['Denominazione']).count()['Sesso']
```

```python
femmine.groupby(['Denominazione']).count()['Sesso']
```

```python
femmine.groupby(['A_NASC']).count()['Sesso']
```

```python
maschi.groupby(['A_NASC']).count()['Sesso']
```

```python
plt.plot(femmine.groupby(['A_NASC']).count()['Sesso'],color='b',label=femmine)
```

```python
plt.plot(maschi.groupby(['A_NASC']).count()['Sesso'], color='r', label='maschi')
```

```python
f = plt.figure()
f.set_figwidth(16)
f.set_figheight(8)
plt.plot(femmine.groupby(['A_NASC']).count()['Sesso'],color='b',label='femmine',marker='.')
plt.plot(maschi.groupby(['A_NASC']).count()['Sesso'], color='r', label='maschi',marker='.')
plt.title('Cittadini di Parma: distribuzione Anno nascita',fontsize=25)
plt.xlabel('Anno di nascita',fontsize=16)
plt.ylabel('numero cittadini',fontsize=16)
xlist=[1910,1920,1930,1940,1950,1960,1970,1980,1990,2000,2010,2020]
xlab=[1910,1920,1930,1940,1950,1960,1970,1980,1990,2000,2010,2020]
plt.xticks(xlist,xlab)
countlist = [y for y in range(0,1750)]
ylist = countlist[::150]
ylab = ylist
plt.yticks(ylist,ylab)
plt.grid(axis='x')
plt.grid(axis='y')
#plt.plot(figsize=(3,4))
plt.legend(loc='lower right',fontsize='large')
plt.show()
```
